# Predicción de valores para un caso hipotético

A partir de un conjunto de datos generados por la herramienta Chat GPT se va a proceder a realizar unas predicciones basadas en los modelos de red neuronal (Total boletos), SVM (Tipo Show) Y Clustering (Clusters).

```python
# Imports basicos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import random

# Load the original dataset
data_path = 'data_with_clusters.csv'
df_original = pd.read_csv(data_path)

# Drop the 'Cluster' column as per the instructions
df_original = df_original.drop(columns=['Cluster'])

# Extract unique values for categorical columns to ensure realistic data generation
categorical_columns = ['Show', 'Nombre Localidad', 'Forma de Pago', 'Estado de Pago', 'EsCompartido', 'TipoShow', 'Mes', 'Día', 'Periodo', 'Mes_Show', 'Día_Show', 'Periodo_Show']
unique_values = {col: df_original[col].unique() for col in categorical_columns}

# Generate a hypothetical dataset for 2025
num_rows = 20000  # Number of rows for the hypothetical dataset
hypothetical_data = {
    'Show': random.choices(list(unique_values['Show'])[:131], k=num_rows),  # Limit to 131 unique values for 'Show'
    'Total Boletos': np.random.randint(1, 10, size=num_rows),  # Random integers for ticket counts
    'Nombre Localidad': random.choices(unique_values['Nombre Localidad'], k=num_rows),
    'Forma de Pago': random.choices(unique_values['Forma de Pago'], k=num_rows),
    'Estado de Pago': random.choices(unique_values['Estado de Pago'], k=num_rows),
    'EsCompartido': random.choices(unique_values['EsCompartido'], k=num_rows),
    'TipoShow': random.choices(unique_values['TipoShow'], k=num_rows),
    'Mes': random.choices(unique_values['Mes'], k=num_rows),
    'Día': random.choices(unique_values['Día'], k=num_rows),
    'Periodo': random.choices(unique_values['Periodo'], k=num_rows),
    'Mes_Show': random.choices(unique_values['Mes_Show'], k=num_rows),
    'Día_Show': random.choices(unique_values['Día_Show'], k=num_rows),
    'Periodo_Show': random.choices(unique_values['Periodo_Show'], k=num_rows)
}

# Create a DataFrame for the hypothetical data
hypothetical_df = pd.DataFrame(hypothetical_data)

# Save the hypothetical dataset to a new CSV file
output_path = 'hypothetical_data_2025.csv'
hypothetical_df.to_csv(output_path, index=False)

print("Hypothetical dataset for 2025 has been generated and saved as hypothetical_data_2025.csv.")
```

```python
import pandas as pd
import random

# Load the previously generated hypothetical dataset
hypothetical_df = pd.read_csv('hypothetical_data_2025.csv')

# Define the desired number of distinct shows and days per month
unique_shows_per_month = {
    'Enero': 25, 'Febrero': 20, 'Marzo': 18, 'Abril': 22, 'Mayo': 15, 'Junio': 19,
    'Julio': 17, 'Agosto': 21, 'Septiembre': 16, 'Octubre': 13, 'Noviembre': 14, 'Diciembre': 12
}
unique_days_per_month = {
    'Enero': 7, 'Febrero': 5, 'Marzo': 6, 'Abril': 7, 'Mayo': 6, 'Junio': 5,
    'Julio': 7, 'Agosto': 6, 'Septiembre': 5, 'Octubre': 6, 'Noviembre': 5, 'Diciembre': 7
}

# Adjust the "Show" and "Días" columns to meet the unique counts per month
adjusted_data = []
for month in unique_shows_per_month.keys():
    # Filter rows for the current month
    month_data = hypothetical_df[hypothetical_df['Mes'] == month]

    # Get a subset of unique shows for the month
    unique_shows = list(hypothetical_df['Show'].unique())[:unique_shows_per_month[month]]
    month_data['Show'] = random.choices(unique_shows, k=len(month_data))

    # Get a subset of unique days for the month
    unique_days = list(hypothetical_df['Día'].unique())[:unique_days_per_month[month]]
    month_data['Día'] = random.choices(unique_days, k=len(month_data))

    # Append the adjusted data for the month
    adjusted_data.append(month_data)

# Combine all adjusted data into a single DataFrame
adjusted_df = pd.concat(adjusted_data)

# Save the adjusted dataset to a new CSV file
adjusted_output_path = 'adjusted_hypothetical_data_2025.csv'
adjusted_df.to_csv(adjusted_output_path, index=False)

print("The dataset has been adjusted to ensure unique show and day counts per month and saved as adjusted_hypothetical_data_with_days_2025.csv.")
```

```python
# Leer el conjunto de datos
df_base = pd.read_csv('adjusted_hypothetical_data_2025.csv')
df_base
```

## Preprocesar los datos

Usaremos los modelos de preprocesamiento para tener el conjunto de datos listo para su predicción.

```python
# Leer el modelo de escalado para Total Boletos
import pickle

with open('totalBoletos_scaler.pkl', 'rb') as file:
    escalador_TotalBoletos = pickle.load(file)

print(escalador_TotalBoletos)
```

```python
df_pred = df_base.copy()

# Escalar Total Boletos de nuestro ejemplo con el modelo escalador importado
df_pred['Total Boletos'] = escalador_TotalBoletos.transform(df_pred[['Total Boletos']])
df_pred
```

```python
import joblib
import json

preprocessor = joblib.load('preprocesador.pkl')

# Cargar los nombres de las columnas transformadas
with open('columns_transformed.json', 'r') as f:
    columns_transformed = json.load(f)

# Preprocesar el ejemplo
df_pred = preprocessor.transform(df_pred)
df_pred = pd.DataFrame(df_pred, columns=columns_transformed)

# Primero, seleccionamos las columnas que NO se deben convertir
columns_to_exclude = ['Show 1', 'Nombre Localidad 1', 'Total Boletos']

# Identificar las columnas que deben ser transformadas a enteros
columns_int = [col for col in df_pred.columns if col not in columns_to_exclude]

# Convertir todas las columnas que no están en 'columns_to_exclude' a enteros
df_pred[columns_int] = df_pred[columns_int].astype(int)
```

```python
# Lista de las columnas que deben ir al inicio en el orden correcto
columnas_inicio = ['Total Boletos', 'Estado de Pago', 'EsCompartido',
                   'Forma de Pago_AppExterna', 'Forma de Pago_PayPhone',
                   'Forma de Pago_Wallet', 'TipoShow_Cine', 'TipoShow_Concierto',
                   'TipoShow_Deporte', 'TipoShow_Donación', 'TipoShow_Ferias',
                   'TipoShow_Otro', 'TipoShow_Talleres', 'TipoShow_Teatro',
                   'Nombre Localidad 1', 'Show 1', 'Día_Domingo', 'Día_Jueves',
                   'Día_Lunes', 'Día_Martes', 'Día_Miércoles', 'Día_Sábado',
                   'Día_Viernes', 'Mes_Abril', 'Mes_Agosto', 'Mes_Enero',
                   'Mes_Febrero', 'Mes_Julio', 'Mes_Junio', 'Mes_Marzo', 'Mes_Mayo',
                   'Mes_Noviembre', 'Mes_Octubre', 'Mes_Septiembre', 'Periodo_Mañana',
                   'Periodo_Noche', 'Periodo_Tarde', 'Día_Show_Domingo', 'Día_Show_Jueves',
                   'Día_Show_Lunes', 'Día_Show_Martes', 'Día_Show_Miércoles', 'Día_Show_Sábado',
                   'Día_Show_Viernes', 'Mes_Show_Abril', 'Mes_Show_Agosto', 'Mes_Show_Diciembre',
                   'Mes_Show_Febrero', 'Mes_Show_Julio', 'Mes_Show_Junio', 'Mes_Show_Marzo',
                   'Mes_Show_Mayo', 'Mes_Show_Noviembre', 'Mes_Show_Octubre', 'Mes_Show_Septiembre',
                   'Periodo_Show_Mañana', 'Periodo_Show_Noche', 'Periodo_Show_Tarde']

# Las columnas que no están en columnas_inicio
columnas_restantes = [col for col in df_pred.columns if col not in columnas_inicio]

# Añadir la columna 'Cluster' al final si está presente en el DataFrame
columnas_restantes.append('Cluster') if 'Cluster' in df_pred.columns else None

# Reorganizar las columnas en el orden deseado
nuevo_orden = columnas_inicio + columnas_restantes

# Reordenar el DataFrame
df_pred_reorganizado = df_pred[nuevo_orden]

# Mostrar el DataFrame con el nuevo orden
df_pred_reorganizado.columns
```

### Prediccion de Cluster

```python
import joblib

# Cargar el modelo KMeans guardado
kmeans_model = joblib.load('kmeans_model.pkl')

# Verificar el tipo de modelo cargado
print(type(kmeans_model))  # Debería ser <class 'sklearn.cluster.KMeans'>
```

```python
df_con_periodo_show_tarde = df_pred_reorganizado.copy()
df_pred_reorganizado = df_pred_reorganizado.drop('Periodo_Show_Tarde', axis=1)
clusters = kmeans_model.predict(df_pred_reorganizado)
```

```python
df_base['Cluster'] = clusters
df_base.head()
```

### Predicción de total boletos

```python
df_base = df_base.drop('Total Boletos', axis=1)
```

```python
# Importar modelo de prediccion de Total Boletos
import tensorflow as tf

# Cargar el modelo desde el archivo .h5
modelo_totalBoletos = tf.keras.models.load_model('modelo_prediccion_totalBoletos.h5')
```

```python
ejemplo_sin_totalBoletos = df_con_periodo_show_tarde.drop(columns=['Total Boletos'])
ejemplo_sin_totalBoletos['Cluster'] = df_base['Cluster']
prediccion_totalBoletos = modelo_totalBoletos.predict(ejemplo_sin_totalBoletos)
```

```python
df_base['Total Boletos Pred'] = prediccion_totalBoletos
# Redondear valores
df_base['Total Boletos Pred'] = df_base['Total Boletos Pred'].round().astype(int)
df_base.head()
```

### Predicción de TipoShow

```python
import joblib

# Cargar el modelo desde el archivo .pkl
modelo_tipoShow = joblib.load('modelo_prediccion_tipoShow.pkl')
```

```python
# Eliminar todas las columnas que contienen 'TipoShow' en su nombre
ejemplo_sin_tipoShow = df_con_periodo_show_tarde.drop(columns=[col for col in df_con_periodo_show_tarde.columns if 'TipoShow' in col])
ejemplo_sin_tipoShow['Cluster'] = df_base['Cluster']

# Ahora puedes usar el modelo para hacer predicciones
prediccion_tipoShow = modelo_tipoShow.predict(ejemplo_sin_tipoShow)
```

```python
df_base = df_base.drop('TipoShow', axis=1)
df_base['TipoShow Pred'] = prediccion_tipoShow
```
